## 1. Import and store the data in a dataframe. (2.5 points)

```python
air_data_set = pd.read_fwf("airq402.dat.txt", header = None) #table of fixed-width formatted (fwf) lines into DataFrame.
```

```python
air_data_set.columns = ["City1", "City2","Avg_Fare", "Distance", "Avg_weekly_passengers", "Mrkt_leading_airline", "Mrkt_share", 
"Avg_fare", "Low_price_airline", "Mrkt_share1", "Price"]
```

```python
air_data_set.info()
```

```python
air_data_set.head()
```

## 2. Remove the outliers from the data (5 points)

```python
q1 = air_data_set.quantile(q=0.25) # for finding 25 percentile
q3 = air_data_set.quantile(q=0.75)  # for finding 75 percentile
IQR = q3 - q1 # calculating Inter quartile range
print(IQR)
```

```python
# finding outliers for each column

High = q3 + 1.5*IQR
Low = q1 - 1.5*IQR
print("High for each column is:\n\n", High)
print("\nLow for each column is:\n\n", Low)
```

```python
new_air_data_set = air_data_set[~((air_data_set<Low)|(air_data_set>High)).any(axis=1)] # removing ouliers from the data set
```

```python
new_air_data_set.shape
```

## 3.Treat “Average Fare” – 3rdColumn as your Dependent Variable and Rest of the columns as Independent Variable. (2.5 points)

```python
new_air_data_set.head()
```

```python
y = new_air_data_set["Avg_Fare"] #target variable
x = new_air_data_set.drop("Avg_Fare", axis =1)  # independent variable
```

## 4. Drop the independent variables which has less than 0.1 correlation with the dependent variable (5 points)

```python
# converting object type columns to categorial type 
new_air_data_set["City1"] = new_air_data_set.City1.astype("category").cat.codes
new_air_data_set["City2"] = new_air_data_set.City2.astype("category").cat.codes
new_air_data_set["Mrkt_leading_airline"] = new_air_data_set.Mrkt_leading_airline.astype("category").cat.codes
new_air_data_set["Low_price_airline"] = new_air_data_set.Low_price_airline.astype("category").cat.codes
```

```python
# finding correlation 
air_data_corr = new_air_data_set.corr()
air_data_corr
```

```python
x = new_air_data_set.drop(["City1","City2","Avg_Fare","Avg_weekly_passengers","Mrkt_leading_airline","Mrkt_share","Low_price_airline","Mrkt_share1"], axis=1) 
# dropped independent variables whose correlation is less than 0.1 from dependent variable
```

## 6. Divide the data set into training and test data set and build a Multiple Linear Regression model. (5 points)

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2) # Spliting data by 80:20 ratio for train and testing respectively
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression_model = LinearRegression()
regression_model.fit(x_train, y_train) # applying linear regression model. Since it has multiple variables it can be called as
## multiple linear regression model
```

## 7. Print the coefficients & intercepts of the linear regression model (5 points)

```python
for idx, col_name in enumerate(x_train.columns):
    print("The coefficient for {} is: {}".format(col_name, regression_model.coef_[idx]))
```

```python
print("The intercept for our model is:",regression_model.intercept_)
```

```python
# we can write our model as y = 6.88 + 0.002*X1 + 0.776*X2 + 0.184*X3 (Coeffecients rounded off to three decimal digits)
```

## 8. Print the accuracy of the overall model (2.5 points)

```python
regression_model.score(x_test,y_test)
```

```python
# So the model explains 97.3% (round-off) of the variability in y using X
```

```python
mse = np.mean((regression_model.predict(x_test)-y_test)**2)
```
