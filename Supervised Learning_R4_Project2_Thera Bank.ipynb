## Get the target column distribution. Your comments

```python
y = bank_data['Personal Loan']
x = bank_data.drop(['ID','Personal Loan'], axis=1)
```

## Split the data into training and test set in the ratio of 70:30 respectively

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import preprocessing
```

```python
x = preprocessing.scale(x)  # scaling the data to increase the overall accuracy since all units will be equal
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

## Use different classification models (Logistic, K-NN and Nave Bayes) to predict the likelihood of a liability customer buying personal loans (15 points)

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
LogR = LogisticRegression()
LogR.fit(x_train, y_train)
y_predict = LogR.predict(x_test)
coef_df = pd.DataFrame(LogR.coef_)
coef_df['intercept'] = LogR.intercept_
print(coef_df)
```

```python
LogR_score = LogR.score(x_test, y_test)
print(LogR_score)
print(metrics.confusion_matrix(y_test, y_predict))
print(metrics.classification_report(y_test, y_predict))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
KNN= KNeighborsClassifier(n_neighbors= 3 , weights = 'distance' )
KNN.fit(x_train, y_train)
y_predict = KNN.predict(x_test)
KNN_score = KNN.score(x_test, y_test)
print(KNN_score)
print(metrics.classification_report(y_test, y_predict))
print(metrics.confusion_matrix(y_test, y_predict))
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

GNB = GaussianNB()
GNB.fit(x_train,y_train)
y_predict = GNB.predict(x_test)
GNB_score = GNB.score(x_test, y_test)
print(GNB_score)
print(metrics.classification_report(y_test, y_predict))
print(metrics.confusion_matrix(y_test, y_predict))
```
